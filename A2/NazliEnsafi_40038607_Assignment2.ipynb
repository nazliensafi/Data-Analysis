#### (1) Quantitative Descriptions 
This folowing block implements a method that takes a dataframe as input, and returns a descriptive dataframe about the input data. It does so by combining series of stats into a single dataframe. 

```python
def quantDDA(df):
    # making sure we're only working with numerical columns
    df = df.select_dtypes(exclude=['object'])
    
    # number of observations
    observations = df.shape[0]
    print("number of observations = ", observations)

    # number of entries (non-empty values)
    entries = df.count(axis = 'rows')
    
    # number of unique entries 
    unique = df.nunique(axis = 'rows')
    
    # number of missing entries
    missing = df.isnull().sum()

    # number of outliers
    Q1 = df.quantile(0.25)
    Q3 = df.quantile(0.75)
    Q2 = df.quantile(0.5)
    IQR = Q3 - Q1
    outliers = ((df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))).sum()
    
    # number of extreme values
    extreme_vals = ((df < (Q1 - 3.0 * IQR)) | (df > (Q3 + 3.0 * IQR))).sum()
    
    # mode 
    mode = df.mode(axis = 'rows', numeric_only = 'True').iloc[0]
    
    # mean 
    mean = df.mean(axis = 'rows', numeric_only = 'True')
    
    # std
    std = df.std(axis = 'rows', numeric_only = 'True')
    
    # max
    maximum = df.max(axis = 'rows', numeric_only = 'True')
    
    # min
    minimum = df.min(axis = 'rows', numeric_only = 'True')
    
    # skewness
    skew = df.skew(axis = 'rows', numeric_only = 'True')
    
    # Kurtosis
    kurt = df.kurtosis(axis = 'rows', numeric_only = 'True')
    
    # create a dictionary
    data = {"entries": entries,
            "unique": unique,
            "missing": missing,
            "outliers": outliers,
            "extreme values": extreme_vals,
            "mode": mode,
            "mean": mean,
            "std": std,
            "max": maximum,
            "min": minimum,
            "skewness": skew,
            "kurtosis": kurt
            }

    # Concatenating the series side
    # by side as depicted by axis=1

    extended_describe = pd.concat(data, axis = 1)
    
    return extended_describe
    
```

```python
df1 = pd.read_csv('tips.csv')
quantDDA(df1)
```
